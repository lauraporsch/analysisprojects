## Importing libraries

```python
import folium
import pandas as pd
```

Creating the base map with focus on Canada (quick google for the "coordinates")

```python
map = folium.Map(
        location=[56.0, -96.0],
        zoom_start=4)

map
```

## Importing and manipulating the dataset to create markers based on provinces

```python
myplaces_df = pd.read_csv('cities_cleaned.csv', index_col=0)
myplaces_df.head()
```

## Creating cluster markers

```python
from folium import plugins
```

```python
cluster = plugins.MarkerCluster().add_to(map)

for lat, long, label in zip(myplaces_df["lat"],myplaces_df["lng"],myplaces_df["city"]):
    folium.Marker(
        location=[lat,long],
        icon=None,
        popup=label).add_to(cluster)
```

```python
map
```

Folium automatically creates Clusters that make sense, based on the geographical closeness. 
Which saves me the time to create single clusters per province.This project is a work in progress, but for now I am happy with it.
